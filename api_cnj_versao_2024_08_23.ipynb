```python
orgao_julgador = 10065

payload = json.dumps({
  "size": 10000,
  "query": {
    "match": {"orgaoJulgador.codigo": orgao_julgador
    #"match": {"orgaoJulgador.codigoMunicipioIBGE": 3538709
    }
  },
  "sort": [{"dataAjuizamento": {"order": "desc"}}] # ou asc
})

headers = {
  'Authorization': api_key,
  'Content-Type': 'application/json'
}

response = requests.request("POST", url, headers=headers, data=payload)
dados = json.loads(response.text)
```

```python
print(dados['hits']['hits'][0])
```

```python
# Lista para armazenar os dados dos processos
processos = []

for processo in dados['hits']['hits']:
    numero_processo = processo['_source']['numeroProcesso']
    classe = processo['_source']['classe']['nome']
    assunto = processo['_source']['assuntos'][0]['nome']  # Somente o primeiro assunto
    data_ajuizamento = processo['_source']['dataAjuizamento']
    ultima_atualizacao = processo['_source']['dataHoraUltimaAtualizacao']
    formato = processo['_source']['formato']['nome']
    codigo = processo['_source']['orgaoJulgador']['codigo']
    orgao_julgador = processo['_source']['orgaoJulgador']['nome']

    # Verificando se a chave 'movimentos' existe e se a lista não está vazia
    if 'movimentos' in processo['_source'] and processo['_source']['movimentos']:
        movimentos = processo['_source']['movimentos']
        # Selecionando o movimento com a data mais recente
        ultimo_movimento = max(movimentos, key=lambda x: x['dataHora'])
        data_ultimo_movimento = ultimo_movimento['dataHora']
        nome_ultimo_movimento = ultimo_movimento['nome']
    else:
        # Se 'movimentos' não existe ou está vazio, definimos os valores como 'None' ou algum valor padrão
        data_ultimo_movimento = None
        nome_ultimo_movimento = None

    # Adicionando os dados na lista de processos
    processos.append([
        numero_processo, classe, assunto, formato, data_ajuizamento, data_ultimo_movimento, ultima_atualizacao,
        nome_ultimo_movimento, codigo, orgao_julgador
    ])

# Criando o DataFrame com as informações dos processos
df = pd.DataFrame(processos, columns=['numero_processo', 'classe', 'assunto', 'formato',
                                      'data_ajuizamento', 'data_ultimo_movimento', 'ultima_atualizacao',
                                      'nome_ultimo_movimento', 'codigo', 'orgao_julgador'])

# Exibindo as 5 primeiras linhas do DataFrame
df.head(5)
```

```python
# Converter as colunas para datetime e ajustar para o horário de São Paulo
df['data_ajuizamento'] = pd.to_datetime(df['data_ajuizamento'], utc=True).dt.tz_convert('America/Sao_Paulo')
df['data_ultimo_movimento'] = pd.to_datetime(df['data_ultimo_movimento'], utc=True).dt.tz_convert('America/Sao_Paulo')
df['ultima_atualizacao'] = pd.to_datetime(df['ultima_atualizacao'], utc=True).dt.tz_convert('America/Sao_Paulo')
df.head(5)
```

```python
# Gravação dos arquivos
df.to_csv('processos.csv', sep=',', header=True, index=False)
```
